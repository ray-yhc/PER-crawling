```python
# BeautifulSoup Special request
# 
# hynix 홈페이지 게시글 크롤링
# 제목, 링크 수집해 표로 만들기
# 
import urllib.request
import bs4
from pandas import Series, DataFrame
import pandas as pd
```

```python
def SinglePage(page):
    result = [] # 결과값은 여기에 저장
    # url에 page번호 붙이기
    url="https://news.skhynix.co.kr/category/Technology/%EB%B0%98%EB%8F%84%EC%B2%B4?page="
    url += str(page)
    
    html=urllib.request.urlopen(url)
    bsObj=bs4.BeautifulSoup(html, "html.parser")
    titles = bsObj.find("div", {"id":"searchList"}).findAll("p",{"class":"searchList_title title"})
    
    # 각 타이틀별로 [제목, 링크] 순으로 저장
    for article in titles:
        artTitle = article.text
        artHref = url + article.find("a")["href"]
        artData = [artTitle, artHref]
        result.append(artData) # [ [제1,링1], [제2,링2], ... ]
    return result

result = [] # 결과값은 여기에 저장
for i in range(1,34):
    print("loading " +str(i)+"th page!!")
    result += SinglePage(i)

fileName="hynix.csv"
table = DataFrame(result)
table.to_csv(fileName, mode="w") # .csv로 저장하기
table
```
